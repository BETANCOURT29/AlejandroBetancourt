# Taller b√°sico pandas - numpy

```python
import pandas as pd
```

```python
# Crear un dataframe que tenga 3 registros(filas) cada una debe contener el nombre de una cancion, el artista que la canta, y la duracion en segundos
data = {
    'cancion': ['Regia', 'Mia', 'Diva'],
    'artista': ['Juan Q', 'Pedro M', 'Marco P'],
    'duracionSeg': [210, 180, 240]
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Mostrar el DataFrame
df
```

```python
# Imprimir la cola del dataframe
df.tail(1)
```

```python
# Acceder a la columna duracion
df.duracionSeg
```

```python
# Calcular el promedio de las duraciones
promedio_duracion = df.duracionSeg.mean()

print(f"El promedio de las duraciones es: {promedio_duracion:.2f} segundos")
```

```python
# Obtener los nombres de las canciones que tengan duracion mayor que el promedio
canciones_mayores_al_promedio = df[df.duracionSeg > promedio_duracion]['cancion']

canciones_mayores_al_promedio
```

```python
# Agregar una columna al dataframe que contenga True si la duracion es mayor que 3 minutos
# y false si es menor
df['mayor_a_3_minutos'] = df.duracionSeg > 180

df
```

```python
# Convertir el dataframe a numpy array
array_numpy = df.to_numpy()

array_numpy
```

```python
# Imprimir el shape del array
print(array_numpy.shape)
```

```python
# Hacer un reshape para convertir el arreglo de 3x3 a un arreglo de 12 elementos
array_reshape = array_numpy.reshape(12)

print(array_reshape)
```
